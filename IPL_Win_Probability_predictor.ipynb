```python
matches= pd.read_csv('matches.csv')
deliveries= pd.read_csv('deliveries.csv')
```

```python
deliveries.shape
```

```python
deliveries.head()
```

```python
matches.head()
```

```python
total_score=deliveries.groupby(["match_id","inning"]).sum()["total_runs"].reset_index()
```

```python
total_score=total_score[total_score["inning"]==1]
```

```python
total_score
```

```python
matches_df = matches.merge(total_score[["match_id",'total_runs']],left_on="id",right_on="match_id")
```

```python
matches_df.head()
```

```python
matches_df["team1"].unique()
```

```python
teams = [ 'Mumbai Indians', 'Kolkata Knight Riders',
       'Rajasthan Royals', 'Chennai Super Kings',
        'Sunrisers Hyderabad',
        'Delhi Capitals', 'Punjab Kings',
       'Lucknow Super Giants', 'Gujarat Titans',
       'Royal Challengers Bengaluru']
```

```python
matches_df["team1"]=matches_df["team1"].str.replace("Delhi Daredevils","Delhi Capitals") #changing names
matches_df["team2"]=matches_df["team2"].str.replace("Delhi Daredevils","Delhi Capitals")
matches_df["team1"]=matches_df["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
matches_df["team2"]=matches_df["team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
matches_df["team1"]=matches_df["team1"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")
matches_df["team2"]=matches_df["team2"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")

```

```python
matches_df=matches_df[matches_df["team1"].isin(teams)]
matches_df=matches_df[matches_df["team2"].isin(teams)]
```

```python
matches_df.shape
```

```python
matches_df.columns
```

```python
matches_df=matches_df[["match_id","city","winner","total_runs"]]
```

```python
matches_df.head()
```

```python
delivery_df = matches_df.merge(deliveries,on="match_id")
```

```python
delivery_df.head()
```

```python
delivery_df= delivery_df[delivery_df["inning"]==2]
```

```python
delivery_df.shape
```

```python
delivery_df.head()
```

```python
delivery_df["total_runs_y"] = delivery_df["total_runs_y"].astype(int)
```

```python
delivery_df.groupby(["match_id"])["total_runs_y"].cumsum()
```

```python
delivery_df["current_score"]=delivery_df.groupby(["match_id"])["total_runs_y"].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df["runs_left"]= delivery_df["total_runs_x"]-delivery_df["current_score"]
```

```python
delivery_df.head()
```

```python
delivery_df["balls_left"]=120-(delivery_df["over"]*6+delivery_df["ball"])
```

```python
delivery_df.columns
```

```python
delivery_df["is_wicket"].dtype
```

```python
delivery_df["wickets_left"]= 10 - delivery_df.groupby(["match_id"])["is_wicket"].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df["crr"]= (delivery_df["current_score"]/(120-delivery_df["balls_left"]))*6
```

```python
delivery_df.head()
```

```python
delivery_df["rr"]= (delivery_df["runs_left"]/(delivery_df["balls_left"]))*6
```

```python
def result (row):
    return 1 if row["batting_team"]==row["winner"] else 0
```

```python
delivery_df["result"]=delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df= delivery_df[["batting_team","bowling_team","city","runs_left","balls_left","wickets_left","total_runs_x","crr","rr","result"]]
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df["balls_left"]!=0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.iloc[:,:-1]
y= final_df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = ColumnTransformer([("onehotencoder",OneHotEncoder(sparse= False , drop="first"),["batting_team","bowling_team","city"])],remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ("step1",onehotencoder),
    ("step2",LogisticRegression(solver="liblinear"))])
```

```python
pipe.fit(X_train  , y_train )
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
